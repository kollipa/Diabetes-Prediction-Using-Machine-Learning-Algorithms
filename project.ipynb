```python
import pandas as pd
import io
df = pd.read_csv(io.StringIO(uploaded['diabetes.csv'].decode('utf-8')))
df.head()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
print(df['Outcome'].value_counts())
print("0: Non-Diabetic\n1: Diabetic")
```

```python
df.groupby('Outcome').mean()
```

```python
X = df.drop(columns = 'Outcome', axis=1)
Y = df['Outcome']
```

```python
X.head()
```

```python
Y.head()
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
Y = df['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data : ', test_data_accuracy)
```

```python
input_data = (5,166,72,19,175,25.8,0.587,51)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
std_data = scaler.transform(input_data_reshaped)
print(std_data)
prediction = classifier.predict(std_data)
print(prediction)
if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
classifier = RandomForestClassifier(n_estimators=100)
classifier.fit(X_train, Y_train)
X_train_prediction = classifier.predict(X_train)
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score of the test data : ', test_data_accuracy)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
classifier = DecisionTreeClassifier()
classifier.fit(X_train, Y_train)
X_train_prediction = classifier.predict(X_train)
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score of the test data : ', test_data_accuracy)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
classifier = MLPClassifier(solver='lbfgs', alpha=0.01, hidden_layer_sizes=(100,))
classifier.fit(X_train, Y_train)
X_train_prediction = classifier.predict(X_train)
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score of the test data : ', test_data_accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC
rf_clf = RandomForestClassifier(n_estimators=100)
svm_clf = SVC(kernel='linear', probability=True)
ensemble = VotingClassifier(estimators=[('rf', rf_clf), ('svm', svm_clf)], voting='hard')
ensemble.fit(X_train, Y_train)
ensemble_prediction = ensemble.predict(X_test)
test_data_accuracy = accuracy_score(ensemble_prediction, Y_test)
print('Accuracy score of the test data using ensemble : ', test_data_accuracy)
```
